```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# Load the dataset
df = pd.read_csv('long_data_.csv')
df.head()
```

```python
# Convert the 'Date' column to datetime
df['Dates'] = pd.to_datetime(df['Dates'])

```

```python
# Set 'Dates' column as the index
df.set_index('Dates', inplace=True)
```

```python
# Sort the DataFrame by date
df.sort_index(inplace=True)
```

```python
# Plot the Power usage over time
plt.figure(figsize=(14, 7))
plt.plot(df['Usage'], label='Use')
plt.title('Power Use Over Time')
plt.xlabel('Date')
plt.ylabel('Use')
plt.legend()
plt.show()
```

```python
# Decompose the time series into trend, seasonality, and residual components
result = seasonal_decompose(df['Usage'], model='additive', period=1)
result.plot()
plt.suptitle('Decomposition of Power Usage')
plt.show()
```
